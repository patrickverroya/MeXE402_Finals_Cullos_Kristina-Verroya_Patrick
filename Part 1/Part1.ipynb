# 3. Demonstrating Morphological Erosion

```python

#3.Demonstrating Morphological Erosion
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Load the input image
img = cv2.imread("OIP.jpg", cv2.IMREAD_GRAYSCALE)

# Check if the image was loaded successfully
if img is None:
    print("Error: Image file not found.")
else:
    # Apply binary thresholding for better visualization of erosion
    _, binary_img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

    # Define different kernel sizes
    kernel_sizes = [(3, 3), (5, 5), (7, 7)]

    # Apply erosion with different kernel sizes
    eroded_images = [binary_img]  # Start with the original binary image
    for size in kernel_sizes:
        kernel = cv2.getStructuringElement(cv2.MORPH_RECT, size)
        eroded = cv2.erode(binary_img, kernel, iterations=1)
        eroded_images.append(eroded)

    # Stack images side by side for comparison
    display = np.hstack(eroded_images)

    # Display the result
    plt.imshow(display, cmap='gray')
    plt.axis('off')  # Hide axes for a cleaner look
    plt.show()
```

# 4. Demonstrating Morphological Dilation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the input image
img = cv2.imread("OIP.jpg", cv2.IMREAD_GRAYSCALE)

# Check if the image was loaded successfully
if img is None:
    print("Error: Image file not found.")
else:
    # Apply binary thresholding for better visualization of dilation
    _, binary_img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

    # Define different kernel sizes
    kernel_sizes = [(3, 3), (5, 5), (7, 7)]

    # Apply dilation with different kernel sizes
    dilated_images = [binary_img]  # Start with the original binary image
    for size in kernel_sizes:
        kernel = cv2.getStructuringElement(cv2.MORPH_RECT, size)
        dilated = cv2.dilate(binary_img, kernel, iterations=1)
        dilated_images.append(dilated)

    # Stack images side by side for comparison
    display = np.hstack(dilated_images)

    # Display the result
    plt.imshow(display, cmap='gray')
    plt.axis('off')  # Hide axes for a cleaner look
    plt.show()

```

# 10. Outlining Shapes with Contours

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread("shapes.png")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

ret, thresh = cv2.threshold(gray,50,255,1)
contours,h = cv2.findContours(thresh,1,2)
# cv2_imshow(thresh)
for cnt in contours:
  approx = cv2.approxPolyDP(cnt,0.01*cv2.arcLength(cnt,True),True)
  n = len(approx)
  if n==6:
    # this is a hexagon
    print("We have a hexagon here")
    cv2.drawContours(img,[cnt],0,255,10)
  elif n==3:
    # this is a triangle
    print("We found a triangle")
    cv2.drawContours(img,[cnt],0,(0,255,0),3)
  elif n>9:
    # this is a circle
    print("We found a circle")
    cv2.drawContours(img,[cnt],0,(0,255,255),3)
  elif n==4:
    # this is a Square
    print("We found a square")
    cv2.drawContours(img,[cnt],0,(255,255,0),3)


plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convert to RGB for Matplotlib
plt.axis('off')
plt.show()
```

# 11. Tracking ball in Video
## In this case I track the DVD Logo

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output, display
import time

# Initialize an empty list to store logo centers
logo_trajectory = []

# Open the video file
cap = cv2.VideoCapture("vid.mp4")
if not cap.isOpened():
    print("Error: Could not open video file.")
    exit()

# Define the video writer
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
out = cv2.VideoWriter(
    'output_dvd_logo.avi',
    cv2.VideoWriter_fourcc(*'MJPG'),
    10,
    (frame_width, frame_height)
)

# Set up matplotlib for real-time frame updates
plt.ion()  # Turn on interactive mode
fig, ax = plt.subplots()
image_display = ax.imshow(np.zeros((frame_height, frame_width, 3), dtype=np.uint8))  # Empty display

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Convert the frame to HSV color space
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Define HSV range for the color of the DVD logo (adjust as needed)
    lower_hue = np.array([0, 50, 50])
    upper_hue = np.array([179, 255, 255])
    mask = cv2.inRange(hsv, lower_hue, upper_hue)

    # Find contours in the mask
    contours, _ = cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    center = None

    if len(contours) > 0:
        # Find the largest contour and its bounding box
        c = max(contours, key=cv2.contourArea)
        x, y, w, h = cv2.boundingRect(c)

        # Filter small contours based on area or width/height thresholds
        if w > 20 and h > 20:  # Adjust thresholds as needed
            center = (int(x + w / 2), int(y + h / 2))
            cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)  # Draw the bounding box
            cv2.circle(frame, center, 5, (0, 255, 0), -1)  # Draw the center point
            logo_trajectory.append(center)

            # Draw the trajectory
            if len(logo_trajectory) > 1:
                for i in range(1, len(logo_trajectory)):
                    if logo_trajectory[i - 1] is not None and logo_trajectory[i] is not None:
                        cv2.line(frame, logo_trajectory[i - 1], logo_trajectory[i], (0, 0, 255), 2)

    # Write the frame to the output video
    out.write(frame)

    # Display the frame using matplotlib
    image_display.set_data(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
    plt.draw()
    plt.pause(0.05)  # Pause to simulate real-time playback

    # Clear previous output and update the frame
    clear_output(wait=True)
    display(fig)

# Release resources
cap.release()
out.release()
plt.ioff()  # Turn off interactive mode
plt.show()  # Show the final plot
```

# 13. Extracting Contours for Shape Analysis

```python
import cv2
import numpy as np

# Load the image
image = cv2.imread('hand.png')  # Replace with your image path
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply GaussianBlur to smooth the image and reduce noise
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Detect edges using Canny edge detection
edges = cv2.Canny(blurred, 50, 150)

# Find contours in the edges image
contours, hierarchy = cv2.findContours(edges.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Loop through the contours and identify shapes
for contour in contours:
    # Approximate the contour to reduce the number of points
    epsilon = 0.04 * cv2.arcLength(contour, True)
    approx = cv2.approxPolyDP(contour, epsilon, True)

    # Draw the contour and the shape
    if len(approx) == 3:
        # Triangle
        cv2.drawContours(image, [approx], -1, (0, 255, 0), 2)  # Green for triangle
    elif len(approx) == 4:
        # Rectangle or square
        x, y, w, h = cv2.boundingRect(approx)
        aspect_ratio = float(w) / h
        if 0.95 <= aspect_ratio <= 1.05:
            cv2.drawContours(image, [approx], -1, (255, 0, 0), 2)  # Blue for square
        else:
            cv2.drawContours(image, [approx], -1, (0, 0, 255), 2)  # Red for rectangle
    elif len(approx) > 4:
        # Circle
        cv2.drawContours(image, [approx], -1, (0, 255, 255), 2)  # Yellow for circle

# Display the original and outlined images
image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray=cv2.cvtColor(gray, cv2.COLOR_BGR2RGB)
display = np.hstack((gray,image))
plt.imshow(display)  # Convert to RGB for Matplotlib
plt.axis('off')
plt.show()


```

# 15. Segmenting Images Based on Contours

```python
import cv2
import numpy as np

# Load the image
image = cv2.imread('forest.jpg')  # Replace 'painting.png' with your image path

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply a binary threshold to highlight contours
_, thresh = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)

# Find contours in the image
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Create a copy of the original image to draw contours
contour_image = image.copy()

# Draw each contour on the image (with a random color for each contour)
for contour in contours:
    # Draw each contour in a random color
    color = np.random.randint(0, 255, (3,)).tolist()
    cv2.drawContours(contour_image, [contour], -1, color, 3)

# Display the original image and the image with contours
image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
contour_image=cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB)
display = np.hstack((image,contour_image))
plt.imshow(display)  # Convert to RGB for Matplotlib
plt.axis('off')
plt.show()
```

# 16. Combining Erosion and Dilation for Feature Refinement

```python
import cv2
import numpy as np

# Read the image
image = cv2.imread('acn.jpg', 0)  # 0 means read as grayscale

# Define a kernel (you can adjust the size based on your needs)
kernel = np.ones((5,20), np.uint8)

# Apply erosion
eroded_image = cv2.erode(image, kernel, iterations=1)

# Apply dilation
dilated_image = cv2.dilate(eroded_image, kernel, iterations=1)

# Show the results
image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
dilated_image=cv2.cvtColor(dilated_image, cv2.COLOR_BGR2RGB)
display = np.hstack((image,dilated_image))
plt.imshow(display)  # Convert to RGB for Matplotlib
plt.axis('off')
plt.show()
```
